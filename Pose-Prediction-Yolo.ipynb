```python
import cv2
from ultralytics import YOLO

# Load the model
model = YOLO("yolov8n-pose.pt")

# Predict on the video
results = model.predict("gesture1.mp4", save=False, show=False, imgsz=320, conf=0.5)

# Define colors for keypoints
color = (255, 0, 0)  # Blue color

# Loop through the predictions in the results
for result in results:
    # Get the processed frame with predictions
    frame = result.plot()

    # Access the keypoints
    keypoints = result.keypoints

    # Iterate over each keypoint
    for i, (x, y) in enumerate(keypoints.xy[0]):
        # Draw a circle around the keypoint
        # cv2.circle(frame, (int(x), int(y)), 3, color, -1)

        # Add text label with keypoint number
        cv2.putText(frame, str(i+1), (int(x) + 5, int(y) + 5), cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 1)

    # Display the frame
    cv2.imshow("Prediction", frame)

    # Wait for 1 ms and check if the user wants to close the window
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the video capture and close all OpenCV windows
cv2.destroyAllWindows()
```

```python
import cv2
from ultralytics import YOLO

# Read the image
image = cv2.imread('man2.jpg')

# Perform inference
results = model(image)

results[0].keypoints

# The keypoints typically come in the form of a list of coordinates
# Assuming results contains the keypoints in the order similar to COCO dataset
keypoints = results[0].keypoints


# Define the keypoints labels in COCO order
keypoint_labels = [
    "Nose", "Left Eye", "Right Eye", "Left Ear", "Right Ear", 
    "Left Shoulder", "Right Shoulder", "Left Elbow", "Right Elbow", 
    "Left Wrist", "Right Wrist", "Left Hip", "Right Hip", 
    "Left Knee", "Right Knee", "Left Ankle", "Right Ankle"
]

# Print the keypoints with their labels
for i, keypoint in enumerate(keypoints[0].xy[0]):
    x, y = int(keypoint[0]), int(keypoint[1])  # Assuming each keypoint has (x, y, confidence)
    print(f"{i}. {keypoint_labels[i]}: (x: {x}, y: {y}, confidence:{keypoints.conf[0][i]:.2f})")
    cv2.circle(image, (int(x), int(y)), 3, (0, 255, 0), -1)
    cv2.putText(image, str(i+1), (int(x) + 5, int(y) + 5), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 1)
    cv2.putText(image, keypoint_labels[i], (int(x) + 5, int(y) + 20), cv2.FONT_HERSHEY_SIMPLEX, 0.3, (0, 0, 255), 1)
    cv2.putText(image, f"{x}-{y}", (int(x) + 5, int(y) + 35), cv2.FONT_HERSHEY_SIMPLEX, 0.3, (255, 0, 0), 1)

cv2.imshow("Keypoints", image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
def verify_arms_up(points):
    head, right_wrist, left_wrist = 0, 0, 0
    head, right_wrist, left_wrist = points[0][1], points[9][1], points[10][1]

    if left_wrist < head and right_wrist < head:
        return True
    else:
        return False
    
    
def verify_legs_up(points):

    left_Leg, right_Leg, left_shoulder, right_shoulder = 0, 0, 0, 0
    left_Leg, right_Leg, left_shoulder, right_shoulder = points[15][0], points[16][0], points[5][0], points[6][0]

    if left_Leg > left_shoulder and right_Leg < right_shoulder:
        return True    
    return False


```

```python
video = 'gesture1.mp4'
capture = cv2.VideoCapture(video)
conneced, frame = capture.read()

keypoint_labels = [
    "Nose", "Left Eye", "Right Eye", "Left Ear", "Right Ear", 
    "Left Shoulder", "Right Shoulder", "Left Elbow", "Right Elbow", 
    "Left Wrist", "Right Wrist", "Left Hip", "Right Hip", 
    "Left Knee", "Right Knee", "Left Ankle", "Right Ankle"
]

threshold = 0.1
while conneced:
    connected, frame = capture.read()

    if not connected:
        break

    # Perform inference
    results = model(frame)

    # The keypoints typically come in the form of a list of coordinates
    # Assuming results contains the keypoints in the order similar to COCO dataset
    keypoints = results[0].keypoints

    # Print the keypoints with their labels
    for i, keypoint in enumerate(keypoints[0].xy[0]):
        x, y = int(keypoint[0]), int(keypoint[1])  # Assuming each keypoint has (x, y, confidence)
        print(f"{i}. {keypoint_labels[i]}: (x: {x}, y: {y}, confidence:{keypoints.conf[0][i]:.2f})")
        cv2.circle(frame, (int(x), int(y)), 3, (0, 255, 0), -1)
        cv2.putText(frame, str(i+1), (int(x) + 5, int(y) + 5), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 1)
        cv2.putText(frame, keypoint_labels[i], (int(x) + 5, int(y) + 20), cv2.FONT_HERSHEY_SIMPLEX, 0.3, (0, 0, 255), 1)
        cv2.putText(frame, f"{x}-{y}", (int(x) + 5, int(y) + 35), cv2.FONT_HERSHEY_SIMPLEX, 0.3, (255, 0, 0), 1)

    if verify_arms_up(keypoints[0].xy[0]) and verify_legs_up(keypoints[0].xy[0]):
        cv2.putText(frame, 'Complete', (50,200), cv2.FONT_HERSHEY_COMPLEX, 3, (0,0,255))


    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

    cv2.imshow('', frame)

cv2.destroyAllWindows()
```
